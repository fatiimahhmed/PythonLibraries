```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

fig = plt.figure(figsize=(4,2), facecolor="lightskyblue", layout='constrained')
fig.suptitle("First Figure")
ax = plt.subplot()
plt.show()
```

```python
#SCATTERRR PLOTTT
fig = plt.figure(figsize=(4,2), facecolor="orange")
fig = plt.scatter([1,2, 4], [6,8,10], marker="*", color="red")
plt.show()
```

```python
plt.figure(figsize=(6, 4))

rollNumber = [3, 9, 18, 14]
marks = [55, 67, 80, 95]

plt.plot(marks, rollNumber, "bo", markersize=5)

plt.xticks(np.arange(50, 101, 10))
plt.yticks(np.arange(0, 21, 5))

plt.xlabel("Marks")
plt.ylabel("Roll Number")

plt.tight_layout()
plt.show()
```

```python
dataset = pd.read_csv(
    "https://raw.githubusercontent.com/MainakRepositor/Datasets/master/Loan%20Eligibility/loan-test.csv"
)

fig = plt.figure(figsize=(8,4), facecolor="lightblue")
plt.suptitle("Loan Amount w.r.t Loan Amount Term")
fig = plt.plot(dataset["LoanAmount"], dataset["Loan_Amount_Term"], "go",alpha=0.5, markersize=5)
# fig = plt.plot(dataset["LoanAmount"], dataset["ApplicantIncome"], "ro",alpha=0.5, markersize=5)

plt.xlabel("Loan Amount")
plt.ylabel("Loan Amount Term")
plt.show()
```

```python
#LINE PLOTTT
rollNum = [12, 34, 45, 56, 78, 89, 90]
marks = [60, 65, 70, 77, 88, 90, 94]

fig = plt.figure(figsize=(8,4))
fig = plt.plot(marks, rollNum, "r-", linewidth=5)
fig = plt.plot(marks, rollNum, "bo", markersize=10)

# --
# -.
# :

plt.xlabel("Marks")
plt.ylabel("Roll Numbers")
plt.show()
```

```python
marks = [60, 65, 70, 77, 84]
colors = ["red", "blue", "orange", "black", "pink"]
subjects = ["english", "urdu", "maths", "islamiat", "computer"]

# fig = plt.barh(subjects, marks, color=colors, edgecolor="black", linestyle="--", linewidth=1)
fig = plt.bar(subjects, marks, color=colors, edgecolor="black", linestyle=":", linewidth=1, width=0.6)
plt.show()
```

```python
marks1 = [86, 89, 90, 92]
marks2 = [67, 77, 80, 90]
subjects = ["english", "urdu", "maths", "computer"]
colors = ["red", "blue", "orange", "black"]
subjects_len = np.arange(len(subjects))
width=0.2

fig = plt.figure(figsize=(12, 4))
fig = plt.bar(subjects_len, marks1, color=colors, width=width)
fig = plt.bar(subjects_len+width, marks2, color=colors, width=width, alpha=0.5)

plt.show()
```

```python
#histtype - bar, barstacked, step, stepfilled

students_50marks = np.random.randint(0, 100, (50))

fig = plt.figure(figsize=(8,8))
fig = plt.xticks(np.arange(0, 100, 5))
fig = plt.hist(students_50marks, color="pink", linewidth=1, edgecolor="black", orientation="vertical", rwidth=0.6)
plt.show()
```

```python
#histtype - bar, barstacked, step, stepfilled

students_50marks1 = np.random.randint(0, 100, (50))
students_50marks2 = np.random.randint(0, 100, (50))

bins = np.arange(0, 100, 5)
fig = plt.figure(figsize=(8,8))
fig = plt.yticks(np.arange(0, 100, 5))
fig = plt.hist([students_50marks1, students_50marks2], color=["pink", "green"], linewidth=1, edgecolor="black", bins=bins, orientation="horizontal")
plt.show()
```
