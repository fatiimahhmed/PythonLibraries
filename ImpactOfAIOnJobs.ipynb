```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_csv("AI_Impact_on_Jobs_2030.csv")

dataset.drop_duplicates()
dataset.dropna()

dssize = dataset.size
dslength = len(dataset)
dscolumns = dssize/dslength

print("Size of Dataset (Rows x Columns)", dslength, "x" , dscolumns)
```

```python
unique_jobs_count = dataset["Job_Title"].nunique()
print("Number of unique job titles:", unique_jobs_count)

top10_jobs = dataset["Job_Title"].value_counts().head(10)
print("\nTop 10 most common job titles:\n", top10_jobs)
```

```python
#Number of Unique skills in dataset
skills_set = dataset.columns[8:19]
all_skills = pd.Series(dataset[skills_set].values.ravel())
unique_skills_count = all_skills.nunique()
print("Number of unique skills:", unique_skills_count)
```

```python
#Count frequency of each skill across all skill columns
skill_counts = dataset[skills_set].stack().value_counts().head(10)

print("Most frequent skills:\n", skill_counts)
```

```python
# Count number of non-null skills per row
dataset['Num_Skills'] = dataset[skills_set].notna().sum(axis=1)

# Find jobs with the highest number of skills listed
top_jobs_by_skills = dataset[['Job_Title', 'Num_Skills']].sort_values(by='Num_Skills', ascending=False)

print("Jobs with the most skills listed:\n", top_jobs_by_skills.head(10))
```

```python
# Group by Job_Title and get mean values of relevant columns
job_stats = dataset.groupby("Job_Title")[["Average_Salary", "AI_Exposure_Index", "Automation_Probability_2030"]].mean()

# Check correlation
correlation = job_stats.corr()
print(correlation)
```

```python
# Salary statistics per job
salary_stats = dataset.groupby("Job_Title")["Average_Salary"].agg(["min", "max", "mean"])
print("Salary statistics by job:\n", salary_stats)
```

```python
# Highest paying jobs (by average salary)
top_salary = salary_stats.sort_values(by="mean", ascending=False).head(10)

plt.figure(figsize=(12, 5))
sns.barplot(x=top_salary.index, y=top_salary["mean"])
plt.title("Top 10 Highest Paying Jobs", fontsize=14)
plt.xlabel("Job Title")
plt.ylabel("Average Salary ($)")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# Lowest paying jobs (by average salary)
lowest_salary_jobs = salary_stats.sort_values(by="mean", ascending=True)
print("\nLowest paying jobs:\n", lowest_salary_jobs.head(10))
```

```python
#Automation Probability by Job - Horizontal Bar
auto_prob = dataset.groupby("Job_Title")["Automation_Probability_2030"].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=auto_prob.values, y=auto_prob.index)
plt.title("Automation Probability by 2030 per Job", fontsize=14)
plt.xlabel("Automation Probability")
plt.ylabel("Job Title")
plt.tight_layout()
plt.show()
```

```python
#AI Exposure Index by Job 
ai_exposure = dataset.groupby("Job_Title")["AI_Exposure_Index"].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=ai_exposure.values, y=ai_exposure.index, hue=ai_exposure, legend=False, palette="Reds_r")
plt.title("Average AI Exposure Index by Job", fontsize=14)
plt.xlabel("AI Exposure Index")
plt.ylabel("Job Title")
plt.tight_layout()
plt.show()
```

```python
# Convert Risk Category into numeric values
risk_category_mapping = {
    "Low": 1,
    "Medium": 2,
    "High": 3
}

plt.figure(figsize=(6, 4))
sns.countplot(data=dataset, x="Risk_Category", order=["Low", "Medium", "High"], hue="Risk_Category", legend=False, palette="Set2")
plt.title("Distribution of Risk Categories", fontsize=14)
plt.xlabel("Risk Category")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
#Salary vs Automation Probability 
plt.figure(figsize=(8, 5))
sns.scatterplot(data=dataset, x="Automation_Probability_2030", y="Average_Salary",
                hue="Risk_Category", palette="Set1", alpha=0.5)
plt.title("Salary vs Automation Probability", fontsize=14)
plt.xlabel("Automation Probability 2030")
plt.ylabel("Average Salary ($)")
plt.tight_layout()
plt.show()
```
