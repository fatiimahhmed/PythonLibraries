```python
import pandas as pd

dataset = pd.read_csv("/content/AI_Impact_on_Jobs_2030.csv")

dataset.drop_duplicates()
dataset.dropna()

dssize = dataset.size
dslength = len(dataset)
dscolumns = dssize/dslength

print("Size of Dataset (Rows x Columns)", dslength, "x" , dscolumns)
```

```python
unique_jobs_count = dataset["Job_Title"].nunique()
print("Number of unique job titles:", unique_jobs_count)

top10_jobs = dataset["Job_Title"].value_counts().head(10)
print("\nTop 10 most common job titles:\n", top10_jobs)
```

```python
#Number of Unique skills in dataset
skills_set = dataset.columns[8:19]
all_skills = pd.Series(dataset[skills_set].values.ravel())
unique_skills_count = all_skills.nunique()
print("Number of unique skills:", unique_skills_count)
```

```python
# Count frequency of each skill across all skill columns
skill_counts = dataset[skills_set].stack().value_counts().head(10)

print("Most frequent skills:\n", skill_counts)
```

```python
# Count number of non-null skills per row
dataset['Num_Skills'] = dataset[skills_set].notna().sum(axis=1)

# Find jobs with the highest number of skills listed
top_jobs_by_skills = dataset[['Job_Title', 'Num_Skills']].sort_values(by='Num_Skills', ascending=False)

print("Jobs with the most skills listed:\n", top_jobs_by_skills.head(10))
```

```python
# Group by Job_Title and get mean values of relevant columns
job_stats = dataset.groupby("Job_Title")[["Average_Salary", "AI_Exposure_Index", "Automation_Probability_2030"]].mean()

# Check correlation
correlation = job_stats.corr()
print(correlation)
```

```python
# Salary statistics per job
salary_stats = dataset.groupby("Job_Title")["Average_Salary"].agg(["min", "max", "mean"])
print("Salary statistics by job:\n", salary_stats)
```

```python
# Highest paying jobs (by average salary)
highest_salary_jobs = salary_stats.sort_values(by="mean", ascending=False)
print("\nHighest paying jobs:\n", highest_salary_jobs.head(10))
```

```python
# Lowest paying jobs (by average salary)
lowest_salary_jobs = salary_stats.sort_values(by="mean", ascending=True)
print("\nLowest paying jobs:\n", lowest_salary_jobs.head(10))
```

```python
# Average AI exposure for each job
avg_ai_exposure = dataset.groupby("Job_Title")["AI_Exposure_Index"].mean()
highest_ai_exposure = avg_ai_exposure.sort_values(ascending=False).head(1)

print("\nJob with highest AI exposure:\n", highest_ai_exposure)


# Job with highest automation probability
highest_auto_prob = dataset.groupby("Job_Title")["Automation_Probability_2030"].mean().sort_values(ascending=False).head(1)

print("\nJob with highest automation probability:\n", highest_auto_prob)
```

```python
# Convert Risk Category into numeric values
risk_category_mapping = {
    "Low": 1,
    "Medium": 2,
    "High": 3
}

dataset["Risk_Category_New"] = dataset["Risk_Category"].map(risk_category_mapping)
print(dataset["Risk_Category_New"].head())
```
