```python
import numpy as np
a = np.array([1, 2, 3, 4])
b = np.array([[1,2], [3, 4]])
c = np.array([[[1,2], [3, 4], [4,5]]])

print(type(a))
print(b)
print(c.dtype)

#ravel -back to single D array, changes the original arrray
c = c.ravel()
print(c)

#flatten -same like ravel, but returns copy
d = c.flatten()
print(d)
```

```python
#size
print(a.size)
print(b.size)
print(c.size)

#shape (rows x column)
print(a.shape)
print(b.shape)
print(c.shape)
```

```python
import matplotlib.pyplot as plt

x_sin = np.arange(0, 2*np.pi, 0.1)
y_sin = np.sin(x_sin)

plt.figure(figsize=(12, 4))
plt.plot(x_sin, y_sin)
plt.title("Sin Curve")
plt.show()
```
