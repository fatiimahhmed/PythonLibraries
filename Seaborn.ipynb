```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

sns.get_dataset_names()

rollNum = [12, 45, 67, 88, 90, 101, 124]
marks = [90, 80, 70, 99, 66, 87, 100]
df = pd.DataFrame({"RollNo":rollNum, "Marks":marks})
print(df.head())

sns.lineplot(x="RollNo", y="Marks", data=df)
plt.title("Student Marks")
```

```python
sds = sns.load_dataset("planets")
sns.set_style("ticks") #whitegrid, ticks, darkgrid
sns.set_context("notebook") #notebook, paper, talk, poster
plt.figure(figsize=(6,6))
sns.scatterplot(data=sds, x=sds["mass"], y=sds["distance"], hue=sds["year"], style=sds["method"], alpha=0.5, palette="Set1") #pastel, Set2, deep, dark, coloblind, muted
sns.despine(left=False)
```

```python
sds = sns.load_dataset("penguins")
sds.head()
sns.set_style("darkgrid") #whitegrid, ticks, darkgrid
plt.figure(figsize=(4,4))
sns.stripplot(data=sds, x=sds["species"], y=sds["body_mass_g"], hue=sds["sex"], palette="dark", dodge=True, jitter=True) 
```

```python
sns.swarmplot(data=sds, x=sds["island"], y=sds["body_mass_g"], hue=sds["species"], palette="dark") 
```

```python
#histogram
# sns.histplot(data=sds, x=sds["body_mass_g"], hue=sds["sex"], multiple="stack") 

#regression plot
plt.figure(figsize=(6,6))
sns.regplot(data=sds, x=sds["bill_depth_mm"], y=sds["bill_length_mm"], scatter=True, color="green") 
```

```python
#barplot
import numpy as np
sns.barplot(data=sds, y=sds["body_mass_g"], x=sds["island"], hue=sds["sex"], palette=["pink", "skyblue"], estimator=np.median)
```

```python
#lineplot
sns.set_style("ticks")
df = pd.read_csv("https://raw.githubusercontent.com/akmadan/seaborn/refs/heads/main/hr_data.csv")
plt.figure(figsize=(6,6))
sns.lineplot(x = "average_montly_hours", y= "number_project", data=df, hue= "left", legend= "full", palette= "flare")
```

```python
#joinplot
sns.jointplot(data=sds, x=sds["body_mass_g"], y=sds["flipper_length_mm"], hue=sds["sex"], kind="hist")

# #countplot
sns.countplot(data=sds, x=sds["species"], hue=sds["sex"] ,palette="Set1")

#boxplot
sns.boxplot(data=sds, x=sds["island"], y=sds["body_mass_g"], hue=sds["sex"] ,palette="Set1")


#violinplot
sns.violinplot(data=sds, x=sds["island"], y=sds["body_mass_g"], hue=sds["sex"] ,palette="Set1", split=True, inner="box")
```

```python
graph = sns.PairGrid(data=sds, hue="sex", palette="Set2")
graph.map_lower(sns.scatterplot)
graph.map_upper(sns.histplot)
graph.map_diag(sns.kdeplot)
```

```python
columns = ["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]
sns.heatmap(data=sds[columns].corr(), annot=True, cmap="coolwarm", vmin=-1, linewidths=0.5, linecolor="black")
```
